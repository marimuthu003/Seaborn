```python
#cs_df=pd.read_csv('ComputerSales.csv')
#cs_df.head()

crash_df=sns.load_dataset('car_crashes')
crash_df
```

## Distribution Plots

### Distribution Plot

```python
sns.displot(crash_df['not_distracted'],kde=False,bins=25)#histogram plots
```

### Joint plot

```python
sns.jointplot(x='speeding',y='alcohol',data=crash_df,kind='hex')#kind=kernel density estimation#kind='hex'/'kde'/'reg'etc.......
```

### KDE Plot

```python
sns.kdeplot(crash_df['alcohol'])
```

 ### Pair Plot

```python
sns.pairplot(crash_df)
```

```python
tips_df=sns.load_dataset('tips')
sns.pairplot(tips_df,hue='sex',palette='Blues')
```

### Rug PLot

```python
sns.rugplot(tips_df['tip'])
```

## Styling

```python
sns.set_style('whitegrid')#/darkgrid/dark/ticks/white

plt.figure(figsize=(8,4))

sns.set_context('paper',font_scale=1.4)

sns.jointplot(x='speeding',y='alcohol',data=crash_df,kind='reg')
 
#sns.despine(left=True,bottom=True,top=True)
```

## Categorical plots

### Bar plots

```python
sns.barplot(x='sex',y='total_bill',data=tips_df,estimator=np.median)
```

### Count plot

```python
sns.countplot(x='sex',data=tips_df)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips_df,hue='sex')
plt.legend(loc=0)
```

### Violin Plot

```python
sns.violinplot(x='day',y='total_bill',data=tips_df,hue='sex',split=True)
plt.legend(loc='upper left')
```

### Strip Plot

```python
plt.figure(figsize=(8,9))
sns.stripplot(x='day',y='total_bill',data=tips_df,hue='sex',jitter=True,dodge=True)#jitter for avoid over plotting and dodge for splitting men and female plots
```

### Swarm plot


```python
sns.swarmplot(x='day',y='total_bill',data=tips_df,hue='sex',dodge=True)
```

### palettes

```python
plt.figure(figsize=(5,5))
sns.set_style('dark')
sns.set_context('paper')
sns.stripplot(x='day',y='total_bill',data=tips_df,hue='sex',jitter=True,dodge=True,palette='twilight')#jitter for avoid over plotting and dodge for splitting men and female plots
```

## Facet Grid


```python
tips_df=sns.load_dataset('tips')
pd.set_option('display.max_rows',244)
tips_df
```

```python
tips_fg=sns.FacetGrid(tips_df,col='time',hue='smoker',height=4,aspect=1.3,col_order=['Dinner','Lunch'],palette='Set1')
tips_fg.map(plt.scatter,'total_bill','tip')
```

```python
kws=dict(s=50,linewidths=.5,edgecolor='w')
tips_fg=sns.FacetGrid(tips_df,col='sex',hue='smoker',height=4,aspect=1.3,hue_order=['yes','no'],hue_kws=dict(marker=['v','x']))
tips_fg.map(plt.scatter,'total_bill','tip',**kws)
plt.show()
```

```python
att_df=sns.load_dataset('attention')
att_fg=sns.FacetGrid(att_df,col='subject',col_wrap=5,height=1.5)
att_fg.map(plt.plot,'solutions','score',marker='.')
```

## Regression plots

```python
plt.figure(figsize=(8,6))
sns.set_context('paper',font_scale=1.4)
sns.lmplot(x='total_bill',y='tip',hue='sex',data=tips_df,markers=['o','x'],scatter_kws={'s':50,'linewidths':0.5,'edgecolor':'w'})
```

```python
sns.set_context('paper',font_scale=1.4)
sns.lmplot(x='total_bill',y='tip',col='sex',row='time',data=tips_df,height=8,aspect=0.65)
```
